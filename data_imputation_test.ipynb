Notebook for data imputation - "cleaner" version, preprocess test data w.r.t. train data

```
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```
# load data
df = pd.read_csv('https://raw.githubusercontent.com/treyra/CS155_miniproject/master/WILDFIRES_TRAIN.csv',sep=',', header=0)
test_df = pd.read_csv('https://raw.githubusercontent.com/treyra/CS155_miniproject/data_visualization_v1/WILDFIRES_TEST.csv',sep=',', header=0)
```

Converting categorical data into numerical data

```
# convert State from strings to numbers
df['STATE'] = df['STATE'].replace(['CA'],1)
df['STATE'] = df['STATE'].replace(['GA'],2)
df.head()
test_df['STATE'] = test_df['STATE'].replace(['CA'],1)
test_df['STATE'] = test_df['STATE'].replace(['GA'],2)
test_df.head()
```

```
# check for unique values of train and test data - do they match
# merge columns to get unique values:
uniqueValues = (df['FIPS_NAME'].append(test_df['FIPS_NAME'])).dropna().unique()
#print(uniqueValues)
```

```
# now replace each unique value (which is not NaN) with a number
for i,fipsname in enumerate(uniqueValues):
  try:
      df['FIPS_NAME'] = df['FIPS_NAME'].replace([str(fipsname)],[i])
  except:
      print('End of unique values for train set')
for i,fipsname in enumerate(uniqueValues):
  try:
      test_df['FIPS_NAME'] = test_df['FIPS_NAME'].replace([str(fipsname)],[i])
  except:
      print('End of unique values for test set')
```

```
test_df.head()
```

```
# same for reporting_source
# merge columns to get unique values:
uniqueValues_rep = (df['SOURCE_REPORTING_UNIT_NAME'].append(test_df['SOURCE_REPORTING_UNIT_NAME'])).dropna().unique()
#print(uniqueValues)
# now replace each unique value (which is not NaN) with a number
for i,repname in enumerate(uniqueValues_rep):
  try:
      df['SOURCE_REPORTING_UNIT_NAME'] = df['SOURCE_REPORTING_UNIT_NAME'].replace([str(repname)],[i])
  except:
      print('End of unique values for train set')
for i,repname in enumerate(uniqueValues_rep):
  try:
      test_df['SOURCE_REPORTING_UNIT_NAME'] = test_df['SOURCE_REPORTING_UNIT_NAME'].replace([str(repname)],[i])
  except:
      print('End of unique values for test set')

  
```

```
# IMPUTE FIPS_CODE and FIPS_NAME FOR TEST SET
# first impute fips_code and fips_name in training dataset
# already done, load here
df_fips_code_imputed =  pd.read_csv('https://raw.githubusercontent.com/treyra/CS155_miniproject/data_visualization_v1/data_cleanup/traindata_fipsname_code_imputed.csv',sep=',', header=0,index_col=0)
# impute the FIPS_CODE for the test data from whole dataset
df_no_labels = df_fips_code_imputed.drop(labels=['LABEL'],axis=1)
frames = [df_no_labels, test_df]
concat_df = pd.concat(frames)
# prepare data to impute code from lat + lon
df_impute_code = concat_df.drop(['id','STATE','DISCOVERY_TIME','FIRE_SIZE','FIPS_NAME','SOURCE_REPORTING_UNIT_NAME','DATE'], axis=1)
from sklearn.impute import KNNImputer
imputer = KNNImputer(n_neighbors=1)
df_code_imputed = imputer.fit_transform(df_impute_code)
```

```
# make pandas frame again from imputed data and include fips_name
df_to_impute_2 = pd.DataFrame({'LATITUDE': df_code_imputed[:, 0], 'LONGITUDE': df_code_imputed[:, 1], 'FIPS_CODE': df_code_imputed[:, 2], 'FIPS_NAME':concat_df['FIPS_NAME']})
# impute fips name
imputer = KNNImputer(n_neighbors=1)
df_code_imputed_2 = imputer.fit_transform(df_to_impute_2)
# merge into test_df
# make test_df_imputed into dataframe again
test_df_imputed = pd.DataFrame({'LATITUDE': df_code_imputed_2[:, 0], 'LONGITUDE': df_code_imputed_2[:, 1], 'FIPS_CODE': df_code_imputed_2[:, 2],'FIPS_NAME': df_code_imputed_2[:, 3]})
# split imputed data by rowindex again
test_df_imputed_2 = test_df_imputed.iloc[len(df):,:]
test_df_imputed_2=test_df_imputed_2.reset_index()
# set imputed values
test_df['FIPS_CODE'] = test_df_imputed_2['FIPS_CODE']
test_df['FIPS_NAME'] = test_df_imputed_2['FIPS_NAME']
test_df.head()
```

```
# convert date into year and day
# convert date column into 2 columns - year and day of the year
import datetime 
dates = pd.DataFrame({'DATE':test_df['DATE']})
#test_df.dtypes
dates['DATE'] = pd.to_datetime(dates['DATE'])
year = np.zeros(len(dates['DATE']))
day = np.zeros(len(dates['DATE']))
for i in range(0,len(dates['DATE'])):
  timetuple = dates['DATE'][:][i].timetuple()
  year[i] = timetuple.tm_year
  day[i] = timetuple.tm_yday
print(year)
print(day)
# add to dataframe
test_df.insert(9,'YEAR',year)
test_df.insert(10,'DAY',day)
test_df = test_df.drop(['DATE'],axis=1)
test_df.head()
```

```
#IMPUTE DISCOVERY TIME
# impute the discovery_time from reporting_source, fips_code, fips_name ??
df_impute_time = concat_df.drop(['id','LATITUDE','LONGITUDE','STATE','FIRE_SIZE','DATE'], axis=1)
# impute discovery time
df_time_imputed = imputer.fit_transform(df_impute_time)
# make dataframe
df_code_imputed_2 = pd.DataFrame({'DISCOVERY_TIME': df_time_imputed[:, 0], 'FIPS_NAME': df_time_imputed[:, 1], 'FIPS_CODE': df_time_imputed[:, 2],'SOURCE_REPORTING_UNIT_NAME': df_time_imputed[:, 3]})
# split imputed data by rowindex again
test_df_imputed = df_code_imputed_2.iloc[len(df):,:]
test_df_imputed = test_df_imputed.reset_index()
# merge back into test_df
test_df['DISCOVERY_TIME']=test_df_imputed['DISCOVERY_TIME']

test_df.head()
```
